```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
X= df.drop('Bankruptcy',axis='columns')
```

```python
figure, ax = plt.subplots(figsize=(20,15))  #Creates empty plot
sns.regplot(data = df,x = ' Net Income to Total Assets',y = " Net worth/Assets", fit_reg = False, color = '#da8600', scatter_kws={'s':df['Bankruptcy']*100}) #Creates Bubble Chart
ax.set_xlabel("Net Income to Total Assets")
ax.set_ylabel("Net worth/Assets")
```

```python
fig = plt.figure(figsize = (15, 8))
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=df[" Research and development expense rate"], ys=df[" Gross Profit to Sales"],zs=df[" Net profit before tax/Paid-in capital"]) #Plot scatter
ax.set_xlabel("Research and development expense rate")
ax.set_ylabel(" Gross Profit to Sales")

ax.set_zlabel("Net profit before tax/Paid-in capital", rotation = 90)
```
